```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from collections import Counter
from pdfminer.high_level import extract_text
import pandas as p
from textblob import TextBlob
```

```python
# Load the book text
# with open("E:/PROJECT/pp/ook.txt", 'r') as f:
#     book_text = f.read()

f=extract_text("E:\PROJECT\pp\Bhagavad-Gita-complete-Parama-Karuna-Devi-version-2016.pdf")
```

```python
# Tokenize the text into sentences
sentences = nltk.sent_tokenize(f)
```

```python
#Tokenize the text into Words
word=nltk.word_tokenize(f)
print(len(word))
print(len(set(word)))
```

```python
# Initialize the sentiment analyzer
#print(len(sentences))
sid = SentimentIntensityAnalyzer()
```

```python
sent=open("Shakuni.txt")
#sentiment = sid.polarity_scores(sent)['compound']
sentim=TextBlob(sent).sentiment
print(sentim)
```

```python
# Identify the main characters
characters = ['Arjuna','sanjay','Shakuni',"gandhari",'Dhritarashtra',"Kripa",'Duryodhana','Bhishma', 'Krishna',"Yudhistira",'Bhima','Draupadi',"Krishna",'Karna','dharmraj',"asvatthama"]  # replace with actual character names
#sample charecters given above
Char_sent={'Arjuna':(),'sanjay':(),'Shakuni':(),"gandhari":(),'Dhritarashtra':(),"Kripa":(),'Duryodhana':(),'Bhishma':(), 'Krishna':(),"Yudhistira":(),'Bhima':(),'Draupadi':(),'Karna':(),"asvatthama":()}
character_counts = Counter()
for sentence in sentences:
    for character in characters:
        #with open('{}.txt'.format(character), 'a') as output
        output=open('{}.txt'.format(character), 'a')  #opening files named after the charecters name
        if character.lower() in sentence.lower():
            character_counts[character] += 1
            Char_sent[character]+=(sentence,) #in dict
            output.write(sentence) #storing the sentences in files
            output.write("\n\n")

            

print(len(Char_sent["Kripa"]))

```

```python
# Determine the sentiments of the characters
character_sentiments = {}
for character in characters:
    character_sentiments[character] = {'positive': 0, 'negative': 0, 'neutral': 0}
for sentence in sentences:
    for character in characters:
        if character.lower() in sentence.lower():
            sentiment = sid.polarity_scores(sentence)['compound']
            if sentiment > 0:
                character_sentiments[character]['positive'] += 1
            elif sentiment < 0:
                character_sentiments[character]['negative'] += 1
            else:
                character_sentiments[character]['neutral'] += 1

print(character_sentiments)
```

```python
# Print the results
print('Character counts:', character_counts)
for character in characters:
    print(character, 'sentiments:', character_sentiments[character])

# pd=p.DataFrame(character_sentiments)
# pd=pd.transpose()

# pd.to_csv("E:\PROJECT\pp\sentdata.csv")
```
